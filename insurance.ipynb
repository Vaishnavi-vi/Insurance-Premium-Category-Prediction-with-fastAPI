```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("C:\\Users\\Dell\\Downloads\\insurance.csv")
dataset.head()
```

```python
df=pd.DataFrame(dataset)
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df["bmi"]=round(df["weight"]/(df["height"]**2),2)
```

```python
#binning 
labels=["Young","adult","middle-aged","senior-citizen"]
bins=[0,18,35,50,100]
df["age_category"]=pd.cut(df["age"],labels=labels,bins=bins)
```

```python
#lifestyle -risk
def lifestyle_risk(series):
    if series["smoker"] and series["bmi"]>30:
        return "high"
    elif series["smoker"] or series["bmi"]>27:
        return "medium"
    else:
        return "low"
    
df["lifestyle_risk"]=df.apply(lifestyle_risk,axis=1)
```

```python
df.head()
```

```python
df["city"].unique()
```

```python
tier_city_1=["Chennai","Mumbai","Hyderabad","Delhi","Pune","Chandigarh","Kolkata","Banglore"]
tier_city_2=["Jaipur","Indore","Kota","Lucknow","Gaya","Jalandar","Mysore"]
def tier_city(city):
    if city in tier_city_1:
        return 1
    if city in tier_city_2:
        return 0
    
df["tier_city"]=df["city"].apply(tier_city)
```

```python
df.head()
```

```python
df.drop(["age","weight","height","smoker","city"],inplace=True,axis=1)
```

```python
df.head()
```

```python
x=df.loc[:,["income_lpa","occupation","bmi","age_category","lifestyle_risk","tier_city"]]
y=df.loc[:,["insurance_premium_category"]]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
#👉 Quick tip: You don’t need to LabelEncode the target (y) for classifiers like RandomForestClassifier, LogisticRegression, etc., because they can handle string labels directly. You’d only encode y if the model strictly requires numbers (e.g., XGBoost
```

```python
categorical_features=["age_category","lifestyle_risk","occupation"]
numerical_features=["bmi","income_lpa","tier_city"]
```

```python
#create column transformer for OHE
preprocessor=ColumnTransformer(transformers=[("cat",OneHotEncoder(),categorical_features),
                                             ("num","passthrough",numerical_features)
                                             ])
```

```python
#create a pipeline
pipeline=Pipeline(steps=[
    ("preprocessor",preprocessor),
    ("classifier",RandomForestClassifier(random_state=42))
])
```

```python
#slit 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
pipeline.fit(x_train,y_train)
```

```python
#predict and evaluate
y_pred_test=pipeline.predict(x_test)
y_pred_train=pipeline.predict(x_train)
```

```python
x_test.sample()
```

```python
from sklearn.metrics import classification_report,accuracy_score
print(accuracy_score(y_test,y_pred_test))
```

```python
print(classification_report(y_test,y_pred_test))
```
